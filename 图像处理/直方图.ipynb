```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt

def show_opencv_img(img):
    im2 = img[:,:,::-1] 
    plt.imshow(im2)

src = cv2.imread("lena.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(src)
```

```python
# 用opencv来做直方图
hist = cv2.calcHist([src], [0], None, [256], [0, 256])
```

```python
# Numpy中直方图计算
hist, bins = np.histogram(src.ravel(), 256, [0, 256])
```

```python
# 另一种numpy方式
hist = np.bincount(src.ravel(), minlength=256)
```

```python
# 绘制直方图，不需要上边的方式
plt.hist(src.ravel(), 256, [0, 256])
```

```python
# 请注意，这里直方图均衡化，图像必须是灰度图像。
# 直方图均衡化
equ = cv2.equalizeHist(src) # 均衡方式1
# 自适应均衡化，参数可选
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl1 = clahe.apply(src)
```
