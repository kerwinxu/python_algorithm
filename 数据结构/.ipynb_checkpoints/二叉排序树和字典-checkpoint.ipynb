# 二叉排序树 (Binary sort Tree)

二叉排序树是一种存储数据的二叉树，将字典的数据存储和查询功能融合在字典里。其基本想法是
- 在二叉树的结点里存储字典中的信息。
- 为二叉树安排好一种字典数据项的存储方式，使得字典的查询等操作可以利用二叉树平均高度远小于树中结点个数的性质，使得检索沿着树中路径进行，从而获得较搞的检索效率。

二叉排序树可以用于实现关键码有序的字典，  
**定义**  
二叉排序树是一种在结点里存储数据的二叉树，一颗二叉树或者为空，或者具备如下性质。
- 其根节点保存着一个数据项（及其关键码）
- 如果其左子树不空，那么其左子树的所有结点保存的（关键码）值均小于根节点保存的（关键码）值。
- 如果其右子树不空，那么其右子树的所有结点保存的（关键码）值均大于根节点保存的（关键码）值。
- 非空的左子树或右子树也是二叉排序树。  

**性质：一颗结点中存储着关键码的二叉树是二叉排序树，当且仅当通过中序遍历这颗二叉树得到的关键码序列是一个递增序列**

```python
# 程序中需要堆栈，我这里实现一个简单的堆栈吧
class KStack:
    def __init__(self):
        self._stack = []
    def is_empty(self):
        return len(self._stack)==0
    def push(self,data):
        self._stack.append(data)
    def pop(self): 
        return self._stack.pop(-1)  # 弹出最后一项啦。
```

如下是这个二叉排序树的增改删查

# 查询

这个查询就是二分法吧，
- 如果要搜索的值比根节点的key值小，就往左结点走，
- 如果要搜索的值比根节点的key值大，就往右结点走，
- 如果要搜索的值等于根节点的key值，就说明找到了。
- 另外没有找到就返回None

#  插入和更改

基于检索插入数据的基本算法是：
- 如果二叉树为空，就直接建立一个包含关键码和关联值的树根结点。
- 否则搜索新结点的插入位置，沿着子节点关系向下。
   - 遇到应该走向左子树而左子树为空，或者遇到右子树而右子树为空时，就是找到了新字典项的插入位置，构造新节点并完成插入。
      - 什么叫应当走向左子树，因为根据二叉排序树定义，就是左子树比根节点小啊。
   - 遇到结点里的关键码等于被检索的关键码，直接替换关联值结束

# 删除

假设已经确定应当删除的结点q，它是父节点p的左子节点（右子节点类似），这时有两种情况。
- q是叶节点，这时只需将父节点p到q的引用设为None，删除就完成了，显然，树中剩下的结点依然构成一颗二叉排序树。
- q不是叶节点，就不能简单删除，要保证余下的还是二叉排序树，这时候就分两种情况。
   - 如果q没有左子节点，这时候只需要将q的右子树直接当作p的左节点就可以了。
   - q有左子树，这时先找到q的左子树的最右结点，设为r，显然它没有右子树，用q的左节点代替q作为p的左子结点，并将q的右子树作为r的右子树。

这里只讨论q的子节点怎么变动，因为在二叉排序树种，q肯定小于p的，且q的一堆子节点都是小于p的，那么只要子节点变动啦。

![二叉排序树的删除](.\字典图片\二叉排序树的删除.png)

# 性质分析

有如下的操作：
- 查询
- 插入和修改
- 删除

查询操作是所有操作共有的，复杂度有赖于这个二叉树的高度，如果这个二叉树构造良好，比如完全二叉树，那么开销就是O(log n)，但如果构造畸形，比如都是左节点到底，那么开销就是O(n)，这时最坏情况的时间复杂度。


# 实现代码

```python
# 我下面做简单的二叉树实现的字典类，分3个层次，分别是：
# 最底层是存储字典的键值对的类，
# 中间的是表示二叉树结点的类，
# 最顶层是二叉树实现的字典类
# 字典的元素，关联，这里将字典里的数据简单的分成2部分，一部分与检索有关的关键码，另一部分称为值
from graphviz import Digraph

class Assoc:
    def __init__(self, key, value):
        self.key=key
        self.value=value
    def __lt__(self, other):
        return self.key < other.key
    def __le__(self, other):
        return self.key < other.key or self.key == other.key
    def __str__(self):
        return "Assoc({0},{1})".format(self.key, self.value)

class KBinTNode:
    """二叉树的结点类，只是保存这个结点的数据和左右子树的结点指针的"""
    def __init__(self,data:Assoc=None, left=None, right=None):
        self.data=data
        self.left=left
        self.right=right

class DictBinTree:
    """二叉树实现的字典类"""
    def __init__(self, root:KBinTNode=None):
        self._root = root         # 根节点。
    def is_empty(self):
        return self._root is None
    def search(self,key):
        bt = self._root            # 从根节点开始
        while bt is not None:      # 不断的遍历查找
            entry = bt.data        # 这里保存的字典的数据。
            if key < entry.key:    # 如果要查找的小于这个值
                bt = bt.left       # 就转向左边的子树
            elif key > entry.key:  # 如果要查找的大于这个值。
                bt = bt.right      # 就转向右边的子树
            else:
                return entry.value # 说明找到需要的值啦
        return None                # 查找不到，就返回None
    def values(self):
        """中序生成所有值的方法"""
        t, s = self._root, KStack() # 初始化
        while t is not None or not s.is_empty():   # 若t不为空，或者栈非空，则进入循环
            while t is not None:                   # 
                s.push(t)                          # 非空节点入栈，向左结点遍历
                t = t.left
            t = s.pop()                            # 弹出栈顶指针 
            yield  t.data.value         # 返回这个结点的数据，这里用重新构造的方式实现深拷贝，别修改原先的啊。
            t = t.right                            # 遍历右结点
    def entries(self):
        """中序生成所有值的方法"""
        t, s = self._root, KStack() # 初始化
        while t is not None or not s.is_empty():   # 若t不为空，或者栈非空，则进入循环
            while t is not None:                   # 
                s.push(t)                          # 非空节点入栈，向左结点遍历
                t = t.left
            t = s.pop()                            # 弹出栈顶指针 
            yield t.data.key, t.data.value         # 返回这个结点的数据，这里用重新构造的方式实现深拷贝，别修改原先的啊。
            t = t.right                            # 遍历右结点
        
    def insert(self, key, value):
        """插入"""
        bt = self._root
        if bt is None:                                  # 如果二叉树为空
            self._root = KBinTNode(Assoc(key, value))   # 就构造一个新的根节点
            return                                      # 做完这个就直接返回啦。
        while True:
            entry = bt.data
            if key < entry.key:                            # 
                if bt.left is None:                        # 遇到应当走向左子树但左子树为空。什么叫应当走向左子树，
                    bt.left = KBinTNode(Assoc(key, value)) # 什么叫应当走向左子树，因为根据二叉排序树定义，就是左子树比根节点小啊。
                    return                                 # 所以在这个key小于树根的时候，就应当在左子树上，但左子树为空啊
                bt = bt.left                               # 这样就是找到了这个位置了。这种方式好像是不断的增加高度的。
            elif key > entry.key:
                if bt.right is None:                       # 同理，应当走向右子树，但右子树为空，就找到了这个位置。
                    bt.right =  KBinTNode(Assoc(key, value)) 
                    return 
                bt = bt.right
            else:
                bt.data.value = value                      # 插入操作同时负责修改啦。
                return
    def delete(self, key):
        p, q = None, self._root   # 位置p作为q的父节点。
        # 先查找这个结点啦
        while q is not None and q.data.key != key:
            p = q                 # p是父节点
            if key < q.data.key:
                q = q.left        # q是不断的往下搜索。
            else:
                q = q.right       # q是不断的往下搜索。
            if q is None:
                return             # 没有找到关键码啊。
        # 运行到这里，表示找到了相应的结点，且q是这个结点，而p是这个结点的父节点，如果有双向链表就好了。
        # 如下先判断情况1吧，q没有左子节点
        if q.left is None:
            if p is None:   # q是根节点啊，
                self._root = q.right
            elif q is p.left:       # 然后根据q是p的左子树还是右子树做相应的操作
                p.left = q.right
            else:
                p.right = q.right
            return 
        # 如下是第二种情况，就是q有左子节点。
        r = q.left                  # 找到这个q的左子树的最右结点
        while r.right is not None:
            r = r.right
        r.right = q.right           #  将q的右子树作为r的右子树
        if p is None:               # q是根节点啊
            self._root = q.left
        elif q is p.left:
            p.left = q.left         # 同样都是将q的左结点当作q的父节点的某个结点。
        else:                       # 不同的是，原先q是p的左节点，就加到左节点上，
            p.right = q.left        # 如果原先是右结点，就加到右结点上。
    def print(self):
        """仅仅是输出数据的"""
        for key, value in self.entries():
            print(key,value)
    def showGraph(self):
        """这个只是显示图形的"""
        g =  Digraph('图片')
        # 这里取得所有的二叉树的边。
        # 这个画图仅仅是得到边就可以了，
        notes=[self._root]   # 初始的话，将顶点加进去
        while len(notes)>0:
            _note = notes.pop(0)  # 我这里是弹出第一项吧，
            # 判断这个点是否有左右结点
            if _note.left is not None:
                g.edge(str(_note.data.key),str(_note.left.data.key))
                notes.append(_note.left)
            if _note.right is not None:
                g.edge(str(_note.data.key),str(_note.right.data.key))
                notes.append(_note.right)
        return g
        
```

```python
# 这里测试一下这个二叉树吧。
# 先随机生成不重复的随机数，我这里当作key
import random
all_num = 100
num = 10
result=random.sample(range(1,all_num),num)
print("要排序的key列表{}:".format(result))
_dict = DictBinTree()
for i in result:
    _dict.insert(i, random.randint(100,1000))

print("中序排列的二叉排序树:{0}".format([k for k,v in _dict.entries() ]))
_dict.showGraph()
# 我这里是随机的数字，但看起来这个图形不怎么好看啊。
# 测试一下删除吧
_dict.delete(result[random.randint(0,num)])
_dict.showGraph()
```

# 最佳二叉排序树

平均检索长度，需要考虑如下两种
- 可能使用树中存在的关键码。这种检索叫检索成功
- 也可能使用树中不存在的关键码。这种叫检索失败

这里用扩充二叉树，其外部结点表示各种失败的情况。  
![二叉排序树及其扩充二叉树](.\字典图片\二叉排序树及其扩充二叉树.png)

如果从0开始标记内部结点和外部结点，那么第i个内部结点是在第i个外部结点和第i+1个外部结点之间。在这种扩充二叉树里，关键码的平均检索长度由下列公式给出：
$$\begin{aligned}E(n)=&\frac{1}{w}\left[\sum\limits_{i=0}^{n-1} p_{i}(l_{i}+1)+\sum\limits_{i=0}^{n}q_{i}l_{i}^{'}\right] \\ & w=\sum\limits_{i=0}^{n-1}p_{i}+\sum\limits_{i=0}^{n}q_{i}\\
&l_{i}是内部结点i的层数，而l_{i}{'}是外部结点i的层数\\
&p_{i}是检索内部结点i的关键码的频率，需要做的比较次数是在结点所在层+1的，这里假设根节点的层数为0\\
&q_{i}是检索关键码属于外部结点i代表的关键码集合的频率，检索到达一个外部结点的比较次数恰好等于该结点所在的层数\\
&\frac{p_{i}}{w}就是检索内部结点i的关键码的概率,\frac{q_{i}}{w}就是检索外部结点i的关键码的概率\\
&p_{i},q_{i}可以看多相应结点的权\end{aligned}$$

最佳排序二叉树就是使得检索的平均比较次数达到最小的二叉排序树，

## 最简单情况

概率相同啦，
$$\frac{p_0}{w}=\cdots=\frac{q_0}{w}=\cdots=\frac{1}{2n+1}$$
在这种情况下：最低的树是最好的吧。


如下是构造方法，基本思想是左右子树的节点数均分，由于二叉树是递归的，构造子树可以采用整个树一样的方法，用递归方法描述最方便。
- 令low=0,high=len(a)-1
- 令m=(high+low)/2
- 把a[m]存入正在构造的二叉排序树的根节点t，递归的
   - 将基于元素片段a[low:m-1]的构造二叉排序树作为t的左子树。
   - 将基于元素片段a[m+1:high]的构造二叉排序树作为t的右子树
   - 片段为空时直接返回None，表示空树

上边算法的时间复杂度为O(n),而最高效的排序算法的时间复杂度为O(n log n)，所以整个构造的的复杂度为O(n log n)

```python
class DictOptBinTree(DictBinTree):
    def __init__(self, seq):
        DictBinTree.__init__(self)
        data = sorted(seq)
        self._root = DictOptBinTree.buildOBT(data,0,len(data)-1)
    @staticmethod
    def buildOBT(data, start, end):
        if start > end:
            return None
        mid = (start + end )// 2
        left = DictOptBinTree.buildOBT(data, start, mid-1)
        right = DictOptBinTree.buildOBT(data, mid+1, end)
        return KBinTNode(Assoc(*data[mid]), left, right)
```

```python
# 测试一下啦
import random
all_num = 100
num = 10
result=random.sample(range(1,all_num),num)
result = [(i,1) for i in result]
print("如下是二叉排序树")
_dict_1 = DictBinTree()
for i in result:
    _dict_1.insert(i[0], i[1])
_dict_1.showGraph()

print("如下是最佳二叉排序树")
_dict_2 = DictOptBinTree(result)
_dict_2.showGraph()
# 这个图形好看多了，
```

## 一般情况的最佳二叉排序树

最佳二叉排序树有一个性质，它的任何子树也是最佳二叉排序树 。 
构造方法：
- 约定记法
   - $T(i,j)表示包含内部结点v_i,\cdots,v_{j-1},和相应外部结点e_i,\cdots,e_{j-1},e_{j}的最佳二叉树。$
      - $E(0,n)表示这颗最佳二叉排序树的权值。$
- 步骤1 ， 首先构造出只包含一个内部结点的最佳二叉排序树，
   - $对每个 i \in [0, n] , 内部结点只有v_i的二叉排序树仅有一颗，它自然是最佳二叉排序树。$
   - $对每个 i 做出 T(i, i+1)，计算出相应的E(i, i+1) = q_i, p_i, q_{i+1},就得到了n颗只包含一个内部结点的最佳二叉排序树$
- $步骤2， 对每个 i \in [0, n-2] ， 构造只包含v_i, v_{i+1}的最佳二叉排序树 T(i, i+2) 。$
   - 以构造T(0,2)为例。这时有两种可能。
      - $以v_0为新树的根，e_0为左子树，而T(1,2)为右子树$
      - $以v_1为新树的根，T(0,1)为左子树，以e_2为右子树。$
- $步骤3， 对每个 i \in [0, n-3] ，构造出内部结点只含有v_i, v_{i+1}, v_{i+2}的最佳二叉排序树 T(i, i+3)$
   - 其实这个根步骤2是同理的。
- $一般步骤 m > 1 , 工作进展到第m步时，(m\le n),对每个 i ( 0 < i < m ) , 前面步骤已经构造出n-i+1颗包含i个内部结点的最佳二叉排序树$
$$ T(0,i),T(1,1+i),\cdots, T(n-i,n)$$
$在第m步要基于它们构造出n-m+1颗包含m个内部结点的最佳二叉排序树$
$$T(0,m),T(1,1+m), \cdots, T(n-m, n)$$，在构造每棵树时，都要考虑所有的组合构造方式，从中选择最佳的一棵树，即带权路径长度最短的那一颗二叉排序树。   
总结一下就是。
- $步骤1，构造出只包含1个内部结点的最佳二叉排序树，T(0,1),T(1,2),\cdots, T(n-1,n)$
- $步骤2，根据上一步的结果，构造出所有只包含2个内部结点的最佳二叉排序树，T(0,2),T(1,3),cdots, T(n-2,n).$
- $\cdots$
- $步骤n，基于前n-1步得到的结果构造出T(0,n)，工作完成。$

### 计算带权路径长度

考虑关键码 $key_i, key_{i+1}, \cdots, key_{j-1}$ 的任一段内部结点 $(0 \le i \le j \le n)$ ,与之对应的内/外结点权值 $q_i, p_i, q_{i+1}, \cdots, p_{j-1}, q_j$ ,可以计算出这段权值交错的分段权值之和为 $W(i,j)=p_i+p_{i+1}+\cdots+p_{j-1}+q_i+q_{i+1}+\cdots+q_{j-1}+q_j$, 显然这种和值根数的结构无关。可以对每段i和j实现算出留用。  
在构造树$T(i,j)时，对于所有的i\le k \le j-1 ,T(i,k)和T(k+1,j)都已经构造好，而且它们的代价应该已知，记作 C(i,k)和C(k+1, j)$    对每个k构造以关键码 $key_k$的内部结点为根，$T(i,k)和T(k+1,j)$为左右子树的候选二叉树，不难确认其权值为 
$$C_k(i,j)=W(i,j)+C(i,k)+C(k+1, j)$$
这里加上$W(i,j)$是因为树中所有结点都增加了一层，高度+1，只要加上这个权值就可以啦。  
新构造的最佳二叉树的代价可以直接从已知的最佳二叉树的代价算出，无需重新基于新树的结构去计算。 
$$ C(i,j)=W(i,j)+min\left\{ C(i,k)+C(k+1, j)\ \ \  |\ \ \  i \le k \le j-1 \right\} $$ 

## 算法设计和实现

对这个构造而言，具体的关键码是什么不重要，只有它们的排列和出现频率在计算中才有意义，
算法的输入是两组描述内部/外部结点出现频率的参数。
- n个元素的参数wp给出内部结点的访问频率。
- n+1个元素的参数wq给出外部结点的访问频率。

算法必须事先计算出所有的内外结点交错的频率序列的分段权和$W(i,j),(0\le i \le j \le n)$，算法进行逐步构造出越来越大的最佳二叉排序树$T(i,j)$,以及计算出的代价 $C(i,j)$,这3组数据都需要记录以支持后续步骤。  
这3组都通过下标i和j访问，比较方便的是3个2维数组来记录。大小都是(n+1)x(n+1)维度。
- r[i][j] 记录构造出最佳子树T(i,j)的根节点的下标。
- c[i][j] 记录子树T(i,j)的代价。
- w[i][j] 表示树中相应的内外交错结点段的权值之和。


```python
def build_opt_tree(wp, wq):
    """一般情况下的二叉排序树wp为内部结点的频率，wq是外部结点的频率"""
    num = len(wp) + 1
    if len(wq) !=  num:
        raise ValueError("失败的频率总数量肯定比成功的频率总数量大于1")
    w = [[0]* num for j in range(num)]  # w为频率序列的分段和
    c = [[0]* num for j in range(num)]  # c为走造出的最佳二叉排序子树。
    r = [[0]* num for j in range(num)]  # 记录构造出的最佳子树T(i,j)的根节点下标。
    
    for i in range(num):  # 计算所有的w[i][j]
        w[i][i] = wq[i]
        for j in range(i+1, num):
            w[i][j] = w[i][j-1] + wp[j-1] + wq[j]
            
    for i in range(0, num-1):    # 直接设置只包含一个结点的树。
        c[i][i+1] = w[i][i+1]
        r[i][i+1] = i
    for m in range(2, num):
        # 计算包含m个内部结点的最佳树（n-m+1颗)
        for i in range(0, num-m):
            k0, j = i, i+m
            wmin = inf
            for k in range(i, j):
                # 在(i,j)里找使得c[i][k]+c[k+1][j]的最小的k
                if c[i][k] + c[k][j] < wmin:
                    wmin = c[i][k] + c[k+1][j]
                    k0 = k
            c[i][j] = w[i][j] + wmin
            r[i][j] = k0
    return c, r
    
```

# 平衡二叉排序树

**定义**：平衡二叉排序树是一类特殊的二叉排序树，它或为空树，或者其左右子树都是平衡二叉排序树（是递归结构），而且其左右子树的高度之差的绝对值不超过1 。  
整棵树的平衡由各节点的平衡刻画，结点平衡可以用一个简单的平衡因子（Balance Factor ， BF）来描述，其取值只有 [-1 , 0 , 1 ] 三个情况。  
在AVL树上的插入操作，前一部分等同于普通二叉排序树的插入即，
- 首先通过检索在树中找到插入位置，
   - 遇到应为左子树但左子树为空。
   - 遇到应为右子树但右子树为空。
- 而后插入新结点。

这时出现一个问题，树原先是平衡的，但在插入结点后可能使得树中的某个局部失衡，应考虑必要的调整。   
**插入后的失衡和调整**  
- 如果在检索插入的过程中：
   - **所有途径结点的BF值都是0**， 那么实际的插入结点不会导致这些结点失衡，只是将他们的BF值修改为-1或者1 
   - **如果途径结点的BF中有不是0的**，那么一定存在一个包含实际插入点的最小非平衡子树，即那颗包含新节点插入位置的，其根节点非0的最小子树。如果插入结点后，这颗子树依然维持平衡，而且其高度不变，那么整棵二叉排序树也将保持平衡。是因为该树的高度不变，在它外边的树结点的BF值也会保持不变。
   
假设插入结点的最小非平衡子树的根节点为a，其左子树较高，   


![图的邻接表表示](./二叉树图片/平衡二叉树示例1.png)  

插入点位置吧  
- 如果插入点在a的右子树，插入之后只需要修改a之下直至插入点路径的上所有结点的BF值，根据a的选择，这些结点的原先的BF值都是0，并将
a的BF值修改为0，由于以a为根的子树高度不变，插入和调整对其他部分没有影响。
- 如果插入点在a的左子树，就就像上边的图(2)，这种情况下必须要设法恢复结点a的平衡。

下面考虑具体的恢复操作：分为4种情况处理
- LL型调整，（a的左子树较高，新节点插入在a的左子树的左子树） 
- LR型调整，（a的左子树较高，新节点插入在a的左子树的右子树） 
- RR型调整，（a的右子树较高，新节点插入在a的右子树的右子树） 
- RL型调整，（a的右子树较高，新节点插入在a的右子树的左子树） 

**LL型失衡的恢复**  : 调整b和a的位置，将b作为根，而a当作右子树。
![](./二叉树图片/平衡二叉树值LL型失衡的恢复.png)

**LR(RL)失衡和调整** :   c提升为这颗子树的根，以b和a分别作为c的左右子树的子结点，c原先的左右结点分别放在b、a上。
![平衡二叉树值LR型失衡的恢复](./二叉树图片/平衡二叉树值LR型失衡的恢复.png)

![二叉树上一系列插入和调整的例子](./二叉树图片/二叉树上一系列插入和调整的例子.png)

插入操作的实现部分：
- 用一个变量a来记录距插入位置最近的平衡因子非0的结点，用pa来记录a的父节点。
   - 如果不存在这种结点，考虑a就是树根。
   - 如果在插入a结点后失衡，a就是失衡位置。
- 修改从a的子结点到新结点路径上各结点的平衡因子。
   - 由于a的定义，这段结点上原先的bf都是0
   - 插入后用一个扫描变量p从a的子结点开始遍历，如果新结点插入在p的左子树，就把p的平衡因子改为1，否则将其改为-1.
- 检查以a为根的子树是否失衡，失衡时做调整。
   - 如果a.bf == 0 ，则插入后不会失衡，修改一下平衡因子并结束。
   - 如果a.bf == 1 : 
      - 如果新结点插入到左子树，就出现了失衡。
         - 新结点在a的左子结点的左子树时做LL调整。
         - 新结点在a的左子结点的右子树时做LR调整。
   - 如果a.bf == -1 :
      - 如果新结点插入到右子树，就出现了失衡。
         - 新结点在a的右子结点的右子树时做RR调整。
         - 新结点在a的右子结点的左子树时做RL调整。

```python
class AVLNode(KBinTNode):
    """结点"""
    def __init__(self, data):
        BinTNode.__init__(self,data)
        self.bf=0
class DictAVL(DictBinTree):
    """平衡二叉树"""
    def __init__(self):
        DictBinTree.__init__(self)
    
    @staticmethod
    def LL(a:KBinTNode, b:KBinTNode):
        a.left = b.right
        b.right = a
        a.bf = b.bf = 0
        return b
    @staticmethod
    def RR(a:KBinTNode, b:KBinTNode):
        a.right = b.left
        b.left = a
        a.bf = b.bf = 0
        return b
    @staticmethod
    def LR(a:KBinTNode, b:KBinTNode):
        c = b.right
        a.left , b.right = c.right, c.left
        c.left, c.right = b, a
        if c.bf == 0:  # c本身就是插入结点
            a.bf = b.bf = 0
        elif c.bf == 1:  # 新节点在c的左子树
            a.bf = -1
            b.bf = 1
        else:           # 新结点在c的右子树
            a.bf = 0
            b.bf = 1 
        c.bf = 0
        return c
    @staticmethod
    def RL(a:KBinTNode, b:KBinTNode):
        c = b.left
        a.right , b.left = c.left, c.right
        c.left, c.right = a, b
        if c.bf == 0:  # c本身就是插入结点
            a.bf = b.bf = 0
        elif c.bf == 1:  # 新节点在c的左子树
            a.bf = 0
            b.bf = -1
        else:           # 新结点在c的右子树
            a.bf = 1
            b.bf = 0 
        c.bf = 0
        return c   
    def insert(self, key, value):
        a = p = self.__root
        if a is None:
            self._root = AVLNode(Assoc(key, value))
            return
        pa = q = None   #
        while p is not None:
            if key == p.data.key:   # 如果存在这个key，表示这个是修改操作啦
                p.data.value = value
                return
            if p.bf != 0:
                pa, a = q, p        # 已知最小非平衡树。q是p的父节点，然后pa，a就是保存的。这里！=0 失衡了。
            q = p                   # 这里保存p结点，在下边的不断的下探。
            if key < p.data.key:
                p = p.left
            else:
                p = p.right
        # q是插入点的父节点，pa，a记录的是最小非平衡树。
        # 插入点不就是，应当遇到子结点但没遇到吗？
        node = AVLNode(Assoc(key, value))
        if key < q.data.eky:
            q.left = node  # 作为左子结点
        else:
            q.right = node 
        # 判断这个新节点是在a的左子树还是右子树
        if key < a.data.key:
            p = b = a.left 
            d = 1
        else:
            p = b = a.right
            d = -1
        # 修改b到新结点路径上各个结点的bf值，b为啊的子结点
        while p != node:
            if key < p.data.key:  # p的左子树增高,key小于p.data.key，说明要插入的key在p的左子树问根节点的子树中。
                p.bf = 1
                p = p.left
            else:                 # p的右子树增高
                p.bf = -1         
                p = p.right
        if a.bf == 0:  # a的原先bf为0，不会失衡
            a.bf = d
            return
        if a.bf == -d:  # 新节点在较低的子树里
            a.bf = 0
            return
        # 结点在较高的子树里，要调整。
        if d == 1:
            if b.bf == 1:
                b = DictAVL.LL(a,b) 
            else:
                b = DictAVL.LR(a,b) 
        else:
            if b.bf == -1:
                b = DictAVL.RR(a,b)
            else:
                b = DictAVL.RL(a,b) 
        if pa is None:         # 原a为树根
            self.__root = b
        else:                  # 原a非树根。新树接到正确位置。
            if pa.left == a:
                pa.left =  b
            else:
                pa.right = b
                        
        
```

# 动态多分枝排序树

## B树

**定义**：一颗m阶B树或者为空，或者具有如下特征：
- 树中分支结点至多有m-1个排序存放的关键码。根结点至少有一个关键码，其他结点至少有 (m-1)/2 个关键码，所有叶节点位于同一层，仅用于表示检索失败，实际上不需要表示（例如可以用空引用表示） 。
- 如果一个分支结点有j个关键码，它就有j+1颗子树，这一个结点中保存的是一个序列 $ < p_0, k_0, p_1, k_1, \cdots , p_{j-1}, k_{j-1}, p_j>$，其中$k_i$为关键码，$p_i$为子结点的引用，而且$k_i$大于$p_i$所引用子树里的关键码，小于$p_{i+1}$所引用子树里的所有关键码。

数据结构嘛，操作无非就是增改删查。这里就是，检索，插入，删除吧。  更改合在检索和插入中了。  
- 检索：
   - 基于关键码从根出发，在遇到的分支里检索关键码，
      - 或者找到成功
      - 或者确定了一颗可能存在被检索关键码的子树，并转入子树继续检索，这样下去。
      - 或者检索到叶节点确定失败。
- 插入：
   - 基于关键码找到应该插入数据的位置p（检索到达叶节点，实际插入的位置p是该叶节点的索引所在的结点）
   - 然后分情况插入。
      - 如果结点p的数据项少于m-1项，就直接在这里按序插入。
      - 否则分裂结点p，将p中关键码和新关键码中较大的一半放入一个新建的结点中，居中的关键码插入p的父节点的相应位置。
         -结点分裂时将关键码插入父节点，也是关键码的插入操作，如果父节点当时已经有m-1个关键码，就需要分裂它并导致将某个关键码插入更上一层的结点，这种分裂可能传播给根节点，根节点的分裂导致这棵树升高一层。
- 删除：
   - 如果p是最下层结点：
      - 如果结点其中关键码多于 (m-1)/2 个，直接删除并结束。
      - 如果p中的关键码不足，首先考虑能否由其兄弟结点调整一些关键码过来，如果能行，考虑平均分配2个结点的关键码，并正确设置位于p的父节点里的分割关键码。
      - 如果p及其兄弟结点的关键码过少无法调整，就将p与其中一个兄弟结点合并，这时位于父节点的分割关键码也要拿到合并后的结点里，合并动作也导致p的父节点里减少一个关键码（相当于删除），这也可能导致父节点与其兄弟结点之间的关键码数据调整，或者合并。
      - 这种结点合并可能一层层的传播，可能一直传播到根节点，如果当时根节点仅有两个子结点而且他们需要合并，就会使得整棵树降低一层。
   - 如果需要删除的关键码（或数据）在上层的分支结点，就先找到其左子树的最右关键码（该关键码一定在最下层结点，设为r),把这个关键码的复制到被删除关键码的位置，随后的工作就像从r原先的位置删除一样，同样有上边所说的各种情况。直接删除/兄弟间调整/结点合并。

总结B树的设计原则：
- 保持树形结构和结点中的关键码有序，用分支结点的关键码为相应子树关键码的区分关键码，保证检索操作能正确进行。
- 保证树中从根到所有叶节点的路径等长，并保证分支节点中关键码的个数在确定的最小范围内变化，并因此保持树结构良好。

```python
#我来做一个B树实现的字典
# 分层次，结点负责保存数据以及子结点，树用来保存树根，以及增改删差的操作。

class BNode():
    """B树的结点"""
    def __init__(self):
        # B树的结点应该包括包含索引和数据的表
        # 这个表保存的是是一个序列，
        # <p_0,k_0,p_1,k_1,...,p_{j-1},k_{j-1},p_k>
        # 其中 p_i 为子结点引用，而k_i为Assoc数据，保存的是key，value        
        self._lst_datas = []
        pass
    def size(self):
        """这个结点保存的数据数量，子结点数量应该是这个数量+1"""
        return (self._lst_datas.size()-1)/2
    def get_index(self, key):
        """根据某个key，返回下标,如果为0，2，4，6的偶数下标，则这个对应的是子结点"""
        for i in range(1, 2*self.size(),2):   # 这样i全是用户的数据啦。
            # 对每一个，只需要判断3种情况，比这个小，等于这个，比这个大。
            if key < self._lst_datas[i]:
                return i - 1
            elif key == self._lst_datas[i]:
                return i
            else:
                pass
        return self.size()-1  # 如果上边的全不是，就返回最后一个啦。
    def is_leaf(self, index):
        """判断某个下标代表的是否是叶子结点"""
        # 我的判断很简单，只是判断值是否为None，我这里的叶子结点就是None
        return self._lst_datas[index] is None
    def is_child_node(self, index):
        """判断某个下标是否是子结点"""
        return index % 2 == 1     # 就是奇数啦
    def is_data(self, index):
        """判断某个下标是否是指向数据的"""
        return index %2 == 0      # 就是偶数啦。
    def get_data(self, index):
        return self._lst_datas[index]
    def insert(self, index, key, value):
        """在指定的位置插入一个数据"""
        # 要分两种情况，一种是初始化后，没有数据啊，
        # 而另一种是，已经找到这个位置了，
        if self.size() == 0:
            self._lst_datas.append(None)
            self._lst_datas.append(Assoc(key, value))
            self._lst_datas.append(None)
        else:
            # 这里假设这个位置就是指向叶子结点的
            self._lst_datas.insert(index+1,Assoc(key, value))
            self._lst_datas.insert(index+2,None)
class BTree():
    """B树"""
    def __init__(self, root:BNode=None):
        self._m = 5         # 我这个是做成5阶B树，这里可以调整。最多4个数据，方便计算吧。
        self._root = root   # 原先的为空值。
        pass
    def get_value(self, key):
        """寻找某个key的value"""
        # 从根节点寻找啦
        _p = self._root
        _i = _p.get_index(key)
        # 循环条件就是，指向的是子结点
        while _p.is_child_node(_i):
            if _p.is_leaf(_i):          # 如果是叶子结点，没找到这个key，就返回None，
                return None
            elif _p.is_data(_i):        # 找到了想要的数据了，
                return _p.get_data(_i).value
            else:
                _p = _p.get_data(_i)    # 最后一种情况就是继续往下子结点查询啦。
                _i = _p.get_index(key)
        return None                     # 默认返回None


    def insert(self, key, value):
        """插入操作"""
        # 首先判断是否是没有树啦
        if self._root is None:
            _node = BNode()
            BNode._lst_datas.append(None)
            BNode._lst_datas.append(Assoc(key, value))
            BNode._lst_datas.append(None)
            return      
        # 找到要插入的结点
        # 从根节点寻找啦
        _p = self._root
        _i = _p.get_index(key)
        # 因为插入操作可能导致父节点更新，这里用一个堆栈来保存路径吧。
        _stack = KStack()
        _stack.push((_p,_i))
        # 循环条件就是，死循环，但里边有退出操作,叶子退出，找到值退出。
        while True:
            if _p.is_leaf(_i):          # 如果是叶子结点，就退出，找到了所要的结点了。
                break 
            elif _p.is_data(_i):        # 这说明这个是更次操作，
                _p.get_data(_i).value = value
                return 
            else:
                _p = _p.get_data(_i)    # 最后一种情况就是继续往下子结点查询啦。
                _i = _p.get_index(key)
                _stack.push((_p,_i))
        # 这里表示找到了相应需要插入的结点了
        # _p为找到需要插入的结点
        # _i为叶子结点的下标
        # _stack里是从根节点到这个结点的路径。
        _stack.pop()      # 弹出最后一项吧，因为这一项就是_p 
        
        # 首先判断这个大小啦
        if _p.size() < self._m -1 :
            # 就在这个结点插入啦。
            _p.insert(_i, key, value)
            pass
        else:
            # 这里要分裂结点p
            # 我这里先插入进去啦。
            _p.insert(_i, key, value)   # 假设阶树为5，然后这里就有5个数据了，分成两部分很好分，
            # 生成一个新结点，
            _node_2 = BNode()
            # 然后将_p的后半部分移动到这个新结点,c语言之类的可以用如下的算。
            # _start = _p._lst_datas.size() -1  # 我是从后边开始删除的
            # _mid = (_p._lst_datas.size-1)/2   # 这里就是_p的中间
            # _end = _mid + 1                   # 这里是 
            # 我这个是python啊，这个list类型，我不断的pop弹出数据就可以啦。
            # 弹出次数，就是这个列表尺寸一半
            for i in range((_p._lst_datas.size()-1)/2):
                _a = _p._lst_datas.pop()
                _node_2._lst_datas.insert(0,_a)
            # 然后弹出来的就是中间结点啦。
            _mid = _p._lst_datas.pop()
            # 要往父节点添加2个信息，一个是这个中间的点，一个是新生成的结点
            _p_parent , _p_parent_i = _stack.pop()              # 先从堆栈种弹出父节点
            _p_parment._lst_datas.insert(_p_parent_i+1,_mid)    # 先插入中间的点
            _p_parment._lst_datas.insert(_p_parent_i+1,_node_2) # 然后插入新建的点
            # 如下就是判断父节点是否也需要分裂了。
            while ((not _stack.is_empty()) and (_p_parent.size() > self._m-1)):
                _p_parent , _p_parent_i = _stack.pop()              # 先从堆栈种弹出父节点
                _p_parment._lst_datas.insert(_p_parent_i+1,_mid)    # 先插入中间的点
                _p_parment._lst_datas.insert(_p_parent_i+1,_node_2) # 然后插入新建的点
            else:
                # 这里要判断是否到顶了，如果到顶了，且root还大，就分裂root
                if _stack.is_empty() and self._root.size > self._m -1 :
                    _node_2 = BNode()
                    for i in range((self._root._lst_datas.size()-1)/2):
                        _a = self._root._lst_datas.pop()
                        _node_2._lst_datas.insert(0,_a)
                    # 然后就是新建一个root
                    _mid = _self._root._lst_datas.pop()
                    _old_root = self._root
                    _node_new_root = BNodede()
                    _node_new_root._lst_datas[0]=_old_root
                    _node_new_root._lst_datas[1]= _mid
                    _node_new_root._lst_datas[2]=_node_2
        
        pass
    def delete(self,key):
        pass
```

# B+树

**定义**： 一颗m阶的B+二叉树或者为空，或者满足下列条件的树。
- 树中每个分支结点至多有m颗子树，除根节点外的分支结点至少有[m/2]颗子树，如果根节点不是叶节点，至少有两颗子树。
- 关键码在结点里排序存放，分支结点里的每一个关键码关联着一颗子树，这个关键码等于所关联子树的根节点的最大关键码，叶节点里的每个关键码关联着一个数据项的存储位置，数据项另行存储。
