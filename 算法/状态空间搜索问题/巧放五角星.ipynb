现有 10 颗五角星，在方格线中放置这些五角
星，每条线上有且仅有两颗，怎样进行巧放（已经
有两颗已经放好了，还剩 8 颗未放）。
项目要求
- 请编制 Python 程序进行求解其他 8 个五
角星该如何放置。
- 请用 Python 程序列出符合要求的 10 颗

五角星摆放的所有可能。  
【输出格式要求】  
为每个解决方案一行，一行内按照字母顺序输出
10 个位置。 

```python
# 如下是初始的数据
a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p, q, r, s, t, u, v, w, x, y = range(25)
Graph = [{a, c, e, f, g},
 {a, b, x, m, l},
 {a, v, w, t, r},
 {d, c, b, v, u},
 {d, e, x, p, o},
 {d, f, h, i, j},
 {g, h, x, s, r},
 {g, i, k, y, l},
 {j, k, x, w, u},
 {j, y, m, n, o},
 {l, n, p, q, r},
 {o, q, s, t, u}]
star_pos=[s, h] 
```

我的算法其实很简单，
- 因为有初始的点star_pos=[s, h] ，我就找这个点上的线，再添加上一个点就是只有2个点啦。
- 如果是多条线，那么下一个点的判断需要多个条件，也就是说这个点不能在别的线上有冲突。
   - 比如这条线上只有一个点，其他4个点都可以作为第二个点，但是这里还有一个条件就是这第二个点同时是别的线的，如果别的线已满，那么这个点就不能作为备选了
   - 我要有一个元组，来保存不可能点，比如一开始设置一个点，我再取一个点，然后其他3个点都是不可能点

```python
def get_next_points(already_pos,no_possible_points):
    """取得下一个点的列表，参数是已有的点"""
    # 我的算法其实就是找某条线上只有一个点的线。我再返回这条线上的其他点，那么就是两个点啦
    for _line in Graph:
        # already_pos 是已经设置的点，
        # 一共5个点，我这里只要找有4个点不在这个已经设置的点中就可以了
        _tmp =tuple((i for i in _line if i not in already_pos))
        if len(_tmp) == 4:  # 先筛选这条线上只有一个点的，也就是余下的点个数为4.
            # 且这个点不能和已有的2点在一条线上
            # 然后再筛选 这里边有些点是不能选的，比如这个点在别的线上已经有2个点了，这个点不能选
            _tmp = tuple((i for i in _tmp if i not in no_possible_points))
            # 然后判断是否有值吧
            if len(_tmp) > 0:
                return _tmp , tuple(_line)+no_possible_points # 返回元组，一个是可能布局的点，而另一个是不可能列表，
            # 如果没有值，就会继续下一次的循环啦。
    return (), no_possible_points
def print2(_pos):
    """这个是将数字转换成字母的"""
    _l = [chr(i+0x61) for i in _pos]  # 无非是整形数加上0x61，就是Graph中对应字符啦。
    print("({0})".format(",".join(_l)))

def wujiaoxing(already_pos,no_possible_points):
    """五角星算法"""
    # 首先判断是否已经排满点了
    if (len(already_pos) >= 10):
        print2(already_pos)
    else:
        # 如果没有排满，就选择下一个点进行递归吧
        _tmp1, _no_possible = get_next_points(already_pos, no_possible_points)
        for i in _tmp1:
            wujiaoxing((i,)+already_pos, _no_possible)
wujiaoxing(tuple(star_pos),(g, h, x, s, r))  # star_pos是初始设定啊，第二个参数是由于这个初始设定，这条线其他的点都不能了。
```
