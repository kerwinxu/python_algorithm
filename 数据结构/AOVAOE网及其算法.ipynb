# AOV网、拓扑排序和拓扑序列

AOV网： 用图中的顶点表示某个有一定规模的“工程”里的不同活动，用图中的边表示各项活动之间的不同顺序，这样的有向图称为顶点活动图(activity on vertex network)   
**拓扑排序和拓扑序列**  
可以把AOV网里的有向边看作一种“顺序”关系，拓扑排序就是问，在一个AOV网里的活动能否排成一种全序。  
定义： 对于给定的AOV网N，如果N中所有顶点能够排成一个线性序列 $$S=v_{i_{0}},v_{i_{1}},\cdots,v_{i_{n-1}}$$  
满足： 如果N中存在从顶点$v_i到顶点v_j$的路径，那么S里$v_i就排在v_j$之前，则S称为N的一个拓扑序列，而构造拓扑序列的操作称为拓扑排序。    
一个AOV网存在拓扑序列，当且仅当它不包含回路（存在回路意味着某些活动的开始要以其自身的完成为先决条件，这种现象称为活动之间的死锁。  
将AOV网N的任一个拓扑序列反向得到的序列，都是N的逆网的一个拓扑序列

## 拓扑排序算法

任何一个无回路的AOV网N都可以做出拓扑序列，  
- 从N中选出一个入度为0的顶点作为序列的下一个顶点。
- 从N网中删除所选顶点及其所有的出边。
- 反复执行上边两部操作，直到已经选出了图中的所有顶点，或者再也找不到入度非0的顶点时算法结束。

如果剩下入度非0的的顶点，说明N中有回路，不存在拓扑序列。

如果要算法实现，肯定不用在图上修改删除啦，简单的方法是做一个入度表，记录数值而已，记录所有顶点的入度，当没删除一个顶点及其出边的时候，就修改这个表的入度数值，另外在删除的过程中判断入度是否为0了，可以用一个队列来保存入度为0的顶点。

```python
def toposort(graph):
    vnum = graph.vertex_num()
    indegree = [0]*vnum  # 入度列表
    toposeq = []
    zerov = -1
    for vi in range(vnum):   # 建立初始的入度表
        for vj, w in graph.out_edges(vi):  # 遍历出边
            indegree[vj] += 1              # 这里是入度表，所以是vj
    for vi in range(vnum):   # 建立初始的0度表
        if indegree[vi] == 0:
            indegree[vi] = zerov 
            zerov = vi       # 其实类似一个链表，zerov是最新的一个0度，而indegree[zerov]的则是下一个0度
    for n in range(vnum):
        if zerov == -1:
            return False   # 不存在拓扑排序
        vi = zerov         # 从0度表弹出顶点
        zerov = indegree[zerov]
        toposeq.append(vi)   # 把一个vi加入拓扑序列
        for vj, w in graph.out_edges(vi):  # 检查这个顶点的出边
            indegree[vj] -= 1    # 对面顶点的入度减一
            if indegree[vj] == 0:  # 判断是否是入度为0的。
                indegree[vj] = zerov
                zerov = vj
    return toposeq
```

# AOE网和关键路径

AOE（activity on edge network) 是另一种常用的带权有向图，抽象的看，AOE网是一种无环的带权有向图。其中：
- 顶点表示事件，有向边表示活动，边上的权值通常表示活动的持续事件。
- 图中的一个顶点表示的事件，
   - 它的入度所表示的活动都已经完成
   - 它的出度所表示的活动可以开始的那个状态，把这一个情况看作事件。 
   


## 关键路径算法

求图$G=(V,E)$里的关键路径，下面假定顶点 $v_0$是$G$中的开始事件，$v_{n-1}$是结束事件，$w(<v_i, v_j>)$是边$<v_i,v_j>$的权。  
首先定义几组变量，用它们记录关键路径中确定的信息：  
- 事件$v_j$的最早可能发生时间$ee[j]$,显然，这一时间要根据在它之前的事件（顶点）和相关的活动（边和初始值）确定，不可能更早发生。
   - $ee[j]$可以递推计算
      - $ee[0]$ = 0 (初始事件总是在时刻0发生）
      - $ee[j] = max\{ee[i]+w(<v_i, v_j>)|<v_i, v_j>\in E\}, 1\le j \le n-1$ ，对每个$j$,只需要考虑以$v_j$为终点的入边集。
- 事件$v_i$的最迟允许发生时间 $le[i]$,事件的发生有可能推迟更晚，但有些事件过迟发生就会延误整个工程的进度。可以根据以知的$ee$值反向的递推计算。
   - $le[n-1]==ee[n-1]$ 最后的一个事件绝不能再延迟。
   - $le[i] = min\{le[j]-w(<v_i,v_j>)|<v_i,v_j>\in E\},0 \le i \le n-1$ ,与上边类似，对每个$i$,只需要考虑以$v_i$为始点的出边集。
- 在这个网络活动中 $a_k=<v_i,v_j>$的最早可能开始时间 $e[k]=ee[i]$,以及它的最迟允许开始时间$l[k]=le[j]-w(<v_i,v_j>)$(只要该活动的实际起始时间不晚于这个时间，就不会拖延整个工程的工期)

活动的集合$A=\{a_k | e[k]=le[k]\}$中的所有活动称为这个AOE网里的的关键活动，因为他们中的任何一个推迟开始，都会延误整个工程的工期，$E-A$的活动为非关键活动，对于非关键活动，$i[k]-le[k]\ne 0$，这个差表示活动$a_k$的时间余量，这个实在不延误整体工期的情况下，活动$a_k$可以推迟的时限，所有完全由关键活动构成的从起始点到终点的路径就是图G中的关键路径，关键路径不止一条，可以同时得到。

### 算法实现

在算法中，我们用两个list来直接表示ee和le，他们都以AOE网中的顶点为下标，其中每项记录对应于一个事件的事件。  
应该注意，为正确生成ee和le数据，计算需要按照一定的顺序进行，对于ee，只有算出了前面制约事件的ee之后，才能进一步计算被制约事件的ee值，对le的顺序刚好相反。不难看到，对ee的可行计算顺序就是AOE网上的拓扑顺序，可以按照顶点的任何拓扑序列计算，而对le的正确计算顺序就是逆拓扑顺序，可以按照拓扑序列的反向计算，因此在正式开始计算之前，需要先得到AOE网的一个拓扑序列。  
算法实现分成如下几步：
- 生成AOE网的一个拓扑序列。
- 生成ee表的值，应该按照拓扑序列的顺序计算。
- 生成le表的值，应该按照拓扑序列的逆序计算。
- 最后，数组e和I可以一起计算。

```python
def critical_path(graph):
    """关键路径"""
    def events_earliest_time(vnum, graph, toposeq):
        """ee最早发生事件"""
        ee = [0] * vnum                        # 初始为0，
        for vi in toposeq:
            for vj, w in graph.out_edges(vi):  # 往后递推
                if ee[vi] + w > ee[vj]:        # 事件j更晚结束
                    ee[vj] = ee[vi] + w        # ee[vj]时间只看入边就可以了。
        return ee
    def events_latest_time(vnum, graph, toposeq, eelast):
        """le最迟发生事件"""
        le = [eelast]*vnum               # 就是根据最后的事件的时间往前递推
        for k in range(vnum-2, -1, -1):  # 逆拓扑顺序，往前递推。
            i = toposeq[k]
            for j, w in graph.out_edges(i):
                if le[j] - w < le[i]:    # 事件i应更早开始
                    le[i] = le[j] - w    # 事件i只看出边，
        return le
    def crt_path(vnum, graph, ee, le):
        """计算关键路径"""
        crt_actions = []
        for i in range(vnum):
            for j, w in graph.out_edges(i):
                if ee[i] == le[j] - w:   # 关键活动，i必须立刻运行
                    crt_actions.append((i, j, ee[i]))
        return crt_actions
    toposeq = toposort(graph)   # 取得拓扑顺序
    if toposeq is None:
        return False            # 不存在拓扑顺序，失败
    vnum = graph.vertex_num()   # 取得顶点数
    ee = events_earliest_time(vnum, graph, toposeq)           # 取得最早发生时间
    le = events_latest_time(vnum, graph, toposeq, ee[vnum-1])  # 取得最迟发生事件
    return crt_path(vnum, graph, ee, le)
```

复杂度分析：  
- 拓扑排序的时间复杂度为O(|V|+|E|)
- 求事件的最早可能时间和最迟时间，需要检查图中的每个顶点和每个顶点边的所有边，各检查一次，时间复杂度为O(|V|+|E|)
因此求关键路径的算法的时间复杂度为O(|V|+|E|)

# AOV和AOE的区别


- 首先，他们都是活动图
- 不同的是：
   - AOV是顶点活动图，边表示各个活动的不同顺序
   - AOE的顶点是各个事件，有向边表示活动
